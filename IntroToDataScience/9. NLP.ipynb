## <font color='pink'>Text Vectorization<font>



Text Vectorization is the process of converting text into a numerical representation.

We can accomplish this using different methods.

1. **Binary Term Frequency** - captures presence (1) or absence (0) of a term in a documnent. 
2. **Bag of Words (BoW) Term Frequency** - captures frequency of term in a document.
3. **Normalized Term Frequency L1** - captures normalized BoW Term Frequency in a document.
4. **Normalized TF-IDF (Term Frequency-Inverse Document Frequency) L2** - captures normalized TD-IDF in a document.

### <font color='white'>Binary Term Frequency<font>

```python
import pandas as pd
```

```python
corpus = ["This is a brown house. This house is big. The street number is 1.",
          "This is a small house. This house has 1 bedroom. The street number is 12.",
          "This dog is brown. This dog likes to play.",
          "The dog is in the bedroom."]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tv = TfidfVectorizer(binary=True, norm=False, use_idf=False, smooth_idf=False, lowercase=True, stop_words='english', token_pattern=r'(?u)\b[A-Za-z]+\b', min_df=1, max_df=1.0, max_features=None, ngram_range=(1, 1))
```

```python
data = pd.DataFrame(tv.fit_transform(corpus).toarray(), columns=tv.get_feature_names())
```

```python
data

#This is a brown house. This house is big. The street number is 1.
```

```python
data = pd.read_csv('SPAM text message 20170820 - Data.csv')
```

```python
data
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(data['Message'],  data['Category'], test_size=0.2)
```

```python
from collections import Counter

print(f"Training class distributions summary: {Counter(Y_train)}")
print(f"Test class distributions summary: {Counter(Y_test)}")
```

```python
#Note: The make_pipeline() method is used to create a pipeline using the provided estimators. 
#Note: We can use it when we want to perform operations step by step on some dataset.
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline

model = make_pipeline(TfidfVectorizer(), MultinomialNB())
model.fit(X_train, Y_train)

y_pred = model.predict(X_test)
```

```python
from imblearn.metrics import classification_report_imbalanced

print(classification_report_imbalanced(Y_test, y_pred))
```

```python
#Note: Undersampling is a technique used to balance uneven datasets by keeping all of the data from the minority class and decreasing the size of data of the majority class.
from imblearn.pipeline import make_pipeline as make_pipeline_imb
from imblearn.under_sampling import RandomUnderSampler

model = make_pipeline_imb(TfidfVectorizer(), RandomUnderSampler(), MultinomialNB())
model.fit(X_train, Y_train)

y_pred = model.predict(X_test)
```

```python
print(classification_report_imbalanced(Y_test, y_pred))
```

## <font color='pink'>Word2Vec<font>



Word2Vec is a state of the art algorithm to generate fixed length distributed vector representations of all the words in a huge corpus. 

The effectiveness of Word2Vec is due two reasons:
1. The use of fixed size vectors, which means the vector size does not depend on the number of unique words in the corpus. 
2. Incorporating semantic information in the vector representations. Word2Vec vectors are highly efficient at grouping similar words together. The algorithm can make strong estimates based on the position of the word in the corpus. For example, ”Kid” and “Child” are similar and hence their vector representation will be very similar.

Word2Vec can be implemented in two architectures — Continuous Bag of Words(CBOW) and Skip-Gram. The main idea of Word2Vec revolves around is predicting the context (outside) words based on a center word or vice versa in a fixed size window.

We will look at two Python libraries for Natural Language Processing: [SpaCy](https://spacy.io/) and [Gensim](https://pypi.org/project/gensim/).

SpaCy is an NLP library designed to have fast performance, and has word embedding models built in, therefore it is perfect for a quick and easy start.
Gensim is a topic modelling library that provides access to Word2Vec and other word embeddings and also allows pre-trained word embeddings to be downloaded online and then loaded.

SpaCy is incredibly convenient to use, as you can use models in 7 languages (English, Polish, German, Spanish, Portuguese, French, Italian and Dutch) in different sizes.
