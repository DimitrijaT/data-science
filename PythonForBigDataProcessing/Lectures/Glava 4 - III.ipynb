# Работа со пивот табели и постапка подели-примени-комбинирај

За работа со пивот табели ќе се користи множеството со основни статистички податоци за фудбалерите од англиската Премиер лига, за натпреварувачката година 2020-2021

```python
import pandas as pd
podatoci = pd.read_csv('EPL_20_21.csv')
podatoci.head()
```

```python
PivotTabela = pd.pivot_table(data = podatoci, index = ['Club'])
PivotTabela.head()
```

```python
PivotTabela = pd.pivot_table(data = podatoci, index = ['Club'],
            values = ['Goals', 'Perc_Passes_Completed'])
PivotTabela.head()
```

```python
import numpy as np
PivotTabela = pd.pivot_table(data = podatoci, index = ['Club'],
            values = ['Goals','Age'], aggfunc={'Goals':np.max, 'Age':np.mean})
PivotTabela.head()
```

```python
PivotTabela = pd.pivot_table(data = podatoci, index = ['Club', 'Nationality'],
            values = ['Age'], aggfunc='count')
PivotTabela.rename(columns = {'Age':'No. of players'}, inplace = True)
PivotTabela.head()
```

```python
PivotTabela = pd.pivot_table(data = podatoci, index = ['Club'],columns=['Nationality'],
            values = ['Age'], aggfunc='count')
PivotTabela.head()
```

# Користење на постапката подели-примени-комбинирај

Со употреба на истото множество за фудбалерите од Премиер Лигата ќе се истражат некои основни примени на постапката подели-примени-комбинирај.

```python
import pandas as pd
podatoci = pd.read_csv('EPL_20_21.csv')
podatoci.head()
```

```python
Grupirani = podatoci.groupby('Club')
Grupirani.groups
```

```python
Grupirani.get_group('Arsenal')
```

```python
Grupirani.Age.agg(['min', 'max', 'mean'])
```

```python
def BroiVozrasni(x):
    Brojach = 0
    for i in x:
        if i >= 30:
            Brojach += 1
    return Brojach        
Grupirani.Age.agg(BroiVozrasni)
```
