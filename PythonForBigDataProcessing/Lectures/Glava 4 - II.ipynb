# Спојување и мешање на податоци

Во овој дел ќе се покажат основни постапки за мешање на податоци од различни табели и спојување на табели

```python
import pandas as pd

Lichni = pd.read_excel('Ispit.xlsx', sheet_name = 0)
Ispit = pd.read_excel('Ispit.xlsx', sheet_name = 1)
Lichni.info()
```

```python
Ispit.info()
```

```python
Lichni.dropna(subset = ['Indeks'], inplace = True)
Ispit.dropna(subset = ['Indeks'], inplace = True)
Lichni.info()
```

```python
VnatreshnoSpoeni = pd.merge(Lichni, Ispit)
VnatreshnoSpoeni.info()
```

```python
LevoSpoeni = pd.merge(Lichni, Ispit, how='left')
LevoSpoeni.info()
```

```python
# Ова е пример за спојување со функцијата join, која што не е покажана во предавањата
SoJoin = Lichni.join(Ispit, lsuffix="_left", rsuffix="_right")
SoJoin.info()
```

```python
SoJoin.head(10)
```

```python
ZalepeniDo = pd.concat([Lichni, Ispit], axis=1)
print(len(Lichni), len(Ispit), len(ZalepeniDo))
ZalepeniDo.info()
```
